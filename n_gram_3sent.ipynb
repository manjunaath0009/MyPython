```python
import nltk
```

```python
nltk.download('brown')
```

```python
sentences = nltk.corpus.brown.sents()
```

```python
len(sentences)
```

```python
sentences[0]
```

```python
n_grams={}
```

```python
for sentence in sentences:
  words=[word for word in sentence if word[0].isalpha()]
  for i in range(len(words) - 1):
    if words[i] not in n_grams:
      n_grams[words[i]] = []
      n_grams[words[i]].append(words[i+1])
```
