```python
from pandas import isnull


def fillAge(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass=='1':
            return 37
        
        elif Pclass=='2':
            return 29
        
        else:
            return 24
    
    else:
        return Age
```

```python
train['Age']=train[['Age','Pclass']].apply(fillAge,axis=1)
```

```python
sns.heatmap(train.isnull(), cmap=['yellow', 'blue'])
```

```python
#Delete/drop a column from the table
train.drop('Cabin',axis=1,inplace=True)
train.head(5)
```

```python
sns.heatmap(train.isnull(), cmap=['yellow', 'blue'])
```

```python
pd.get_dummies(train['Embarked'], drop_first=True)
```

```python
sex=pd.get_dummies(train['Sex'],drop_first=True)
embarked=pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
train.drop(['Name','Embarked','Ticket', 'Sex'],axis=1, inplace=True)
```

```python
train.head(5)
```

```python
train=pd.concat([train,sex,embarked],axis=1)
```
