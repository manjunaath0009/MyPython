```python
from cmath import rect
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
# plt.show()
%matplotlib inline 

import numpy as np 

x=np.arange(0,10)
y=np.arange(11,21)

#plt.scatter(x,y,c="g")
plt.xlabel("X Axis",c='r')
plt.ylabel("Y Axis",c='r')
plt.title("First Plot diagram of Manjunaath")
plt.savefig("First_Plot_Manjunaath.jpg")
y=x*x
#plt.plot(x,y,'ro--')s
plt.plot(x,y,linestyle='dashed',linewidth='2',markersize=10)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b')
plt.subplot(2,2,2)
plt.plot(x,y,'r')
plt.subplot(2,2,3)
plt.plot(x,y,'k')
```

```python
# Box Plot
data=np.array([0,5,6,20,15])

plt.boxplot(data,vert="None", patch_artist=True)
```

```python
# Box Plot
data=[np.random.normal(0,std,100) for std in range (1,4)]
plt.boxplot(data,vert="None", patch_artist=True)
data
```

```python
# Pie Chart
labels="Python", "C++", "Ruby", "Java"
sizes=[200,300,100,250]
colors=["gold", "red", "blue", "green"]
explode=(0,0,0,0)
plt.pie(sizes, labels=labels, explode=explode, colors=colors, shadow=True, autopct='%1.1f%%')
plt.axis('equal')
```
