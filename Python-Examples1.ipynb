```python
def winner(erica, bob):
    def calculate_score(participant):
        score = 0
        for c in participant:
            if c == 'E':
                score += 1
            elif c == 'M':
                score += 3
            elif c == 'H':
                score += 5
        return score

    erica_score = calculate_score(erica)
    bob_score = calculate_score(bob)

    if erica_score > bob_score:
        return "Erica"
    elif bob_score > erica_score:
        return "Bob"
    else:
        return "Tie"

if __name__ == '__main__':
    erica = input()
    bob = input()
    result = winner(erica, bob)
    print(result)

```

```python
import matplotlib.pyplot as plt
from sklearn.inspection import PartialDependenceDisplay
from sklearn.ensemble import RandomForestRegressor

n_samples = 500
rng = np.random.RandomState(0)
X = rng.randn(n_samples, 2)
noise = rng.normal(loc=0.0, scale=0.01, size=n_samples)
y = 5 * X[:, 0] + np.sin(10 * np.pi * X[:, 0]) - noise

rf_no_cst = RandomForestRegressor().fit(X, y)
rf_cst = RandomForestRegressor(monotonic_cst=[1, 0]).fit(X, y)

disp = PartialDependenceDisplay.from_estimator(
    rf_no_cst,
    X,
    features=[0],
    feature_names=["feature 0"],
    line_kw={"linewidth": 4, "label": "unconstrained", "color": "tab:blue"},
)
PartialDependenceDisplay.from_estimator(
    rf_cst,
    X,
    features=[0],
    line_kw={"linewidth": 4, "label": "constrained", "color": "tab:orange"},
    ax=disp.axes_,
)
disp.axes_[0, 0].plot(
    X[:, 0], y, "o", alpha=0.5, zorder=-1, label="samples", color="tab:green"
)
disp.axes_[0, 0].set_ylim(-3, 3)
disp.axes_[0, 0].set_xlim(-1, 1)
disp.axes_[0, 0].legend()
plt.show()
```

```python
print("***********C L A S S E S, INHERITENCE, ITERATORS, POLYMORPHISM*************")

class myClass:
    def __init__(self, fname, lname):
        self.fname = fname
        self.lname = lname
    
    def __str__(self):
        return f"{self.fname} {self.lname}"

fullName = myClass('SG', 'Manjunaath')
print(fullName)


class myCLass:
    def __init__(self, fname, lname):
        self.fname = fname
        self.lname = lname
    
    def myClassFun(self):
        print(f"Hello My Full Name is " + self.fname + " " + self.lname)
        

fullName=myCLass("Sai", "Samarth")

fullName.myClassFun()

class mySchool:
    def __init__(self, city, ctry):
        self.city = city
        self.ctry = ctry
        
    def printMySchool(self):
        print(self.city + self.ctry)

x=mySchool("Bangalore", "India")
x.printMySchool()

class personProfile:
    def __init__(self, fname, lname):
        self.fname = fname
        self.lname = lname
    
    def printPerson(self):
        print(self.fname + self.lname)
    
class employeeProfile(personProfile):
    def __init__(self, fname, lname):
        personProfile.__init__(self, fname, lname)  

x=employeeProfile("Hello", " sWorld")
x.printPerson()

class Person:
    def __init__(self, grade, college):
        self.Grade = grade
        self.College = college
    
    def Employee(self):
        print(self.Grade + self.College)
    
class Employee(Person):
    def __init__(self, grade, college, year):
        super().__init__(grade, college)
        self.Year = year
    
    def Welcome(self):
        print(self.Year)

x=Employee("Sai", "Samarth", "2000")
x.Welcome()
x.Employee()
        
class MyNumbers:
    def __iter__(self):
        self.a = 1
        return self
    
    def __next__(self):
        if self.a < 20:
            x=self.a
            self.a +=2
            return x
        else:
            raise StopIteration

myClass=MyNumbers()
myIter=iter(myClass)

for x in myIter:
    print(x, end=" ")


class Cricket:
    def __init__(self, batsman, bowler):
        self.Batsman = batsman
        self.Bowler = bowler
    
    def test(self):
        print("Cricket : " + self.Batsman + self.Bowler)

class Football:
    def __init__(self, goally, forward):
        self.Goally = goally
        self.Forward = forward
    
    def test(self):
        print("Football : " + self.Goally + self.Forward)

print("\n")
cricket=Cricket("A"," B")
football=Football("C"," D")

for x in (cricket, football):
    x.test()


class City1:
    def __init__(self, male, female):
        self.Male=male
        self.Female=female
    
    def printCity(self):
        print("City : " + self.Male + ":" + self.Female)
    

class City2(City1):
    def printCity(self):
        print("City2 : " + self.Male + ":" + self.Female)

city1=City1("100", "200")
city2=City2("300", "400")

for x in (city1, city2):
    x.printCity()
    
```

```python
s="sgm"
print(s[1])
# Code source: Jaques Grobler
# License: BSD 3 clause

import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test, color="black")
plt.plot(diabetes_X_test, diabetes_y_pred, color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
